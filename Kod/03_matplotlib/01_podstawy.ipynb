### Import biblioteki Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

Pierwszy wykres

```python
_ = plt.plot([0, 1], [0,
                      1])  # _ = dodaliśmy aby pozbyć się opisu dodatkowego nad wykresem, [<matplotlib.lines.Line2D at 0x2d78aa80f70>]
```

```python
x = np.arange(-1, 1.1, 0.1)
y = x ** 2
```

```python
_ = plt.plot(x, y)
```

Tytuł wykresu

```python
plt.plot(x, y)
_ = plt.title('Wykres x * x')
```

Dwie funkcje na jednym wykresie

```python
x1 = np.arange(-1, 1.1, 0.1)
y1 = x1 ** 3
```

```python
plt.plot(x, y)
_ = plt.plot(x1, y1)
```

Etykiety

```python
plt.plot(x, y, label='x^2')
plt.plot(x1, y1, label='x^3')
_ = plt.legend()
```

Wiele wykresów

```python
plt.subplot(1, 2, 1)  # subplot(1,2, 1) 1 -wiersz 2 kolumny, numer wykresu 1
plt.plot(x, y)
plt.subplot(1, 2, 2)
_ = plt.plot(x1, y1)
```

Wykres słupkowy

```python
x2 = np.arange(1, 6)
y2 = [10, 20, 40, 15, 99]
```

```python
_ = plt.bar(x2, y2)  #bar jako wykres słupkowy
```

Wykres rozproszony

```python
xs = np.random.rand(500)
ys = np.random.rand(500)
```

```python
_ = plt.scatter(xs, ys)
```

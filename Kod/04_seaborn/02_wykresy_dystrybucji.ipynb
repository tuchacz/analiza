```python
import seaborn as sns
```

Displot - wykres dystrybucji

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
_ = sns.displot(tips, x='tip')
```

```python
_ = sns.displot(tips, x='tip', bins=30)
```

Joinplot - połaczenie relacyjnego z histogramem

```python
_ = sns.jointplot(tips, x='total_bill', y='tip')
```

```python
_ = sns.jointplot(tips, x='total_bill', y='tip', hue='smoker')  # hue pokreśla na kolor dana kolumne
```

Pairplot - na jednym wykresie relacje wszystkich kolumn, porównanie ich

```python
_ = sns.pairplot(tips)
```

```python
_ = sns.pairplot(tips, hue='time')
```

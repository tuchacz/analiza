Import biblioteki seaborn

```python
import seaborn as sns
```

### Scatter - wykres rozrzutu

```python
tips = sns.load_dataset('tips') # pobiera zbior danych
tips.head()
```

```python
_ = sns.relplot(tips, x='total_bill', y='tip') #wykres relacyjny relplot
```

```python
_ = sns.relplot(tips, x='total_bill', y='tip', hue='sex')
```

```python
_ = sns.relplot(tips, x='total_bill', y='tip', hue='sex', size='size')
```

### Line - wykres linii

```python
ice = sns.load_dataset('seaice')
ice.head()
```

```python
ice.info() # informacje
```

```python
_ = sns.relplot(ice[ice['Date'].dt.year == 2018], x='Date', y='Extent', kind='line') #dt.year filtracja na rok 2018, zmiana ze scatter na line kind
```

```python
ice['mm-dd'] = ice['Date'].dt.strftime('%m-%d') # dt datatime - kolumna
```

```python
ice.head()
```

```python
ice_2000_2018 = ice[(ice['Date'].dt.year > 1999) & (ice['Date'].dt.year < 2019)]
```

```python
_ = sns.relplot(ice_2000_2018, x='mm-dd', y='Extent', kind='line')
```

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

Stripplot

```python
_ = sns.catplot(tips, x='day', y='total_bill') # cała grupa wykresów catplot, on grupuje wszystkie wykresy wtedy nie okresla sie kind ktore jest domyslne jako stripplot
```

Swarmplot - pokazuje gdzie jest duze zageszczenie

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='swarm') #definiuje sie kind jako swarm
```

Boxplot - wizualizacja dystrybucji

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='box')  # w boksie jest mediana 50 ponizej od 25 do 50 powyzej 50 do 75, wasy ekstrema, poza punkty wychodzace
```

Bar

```python
_ = sns.catplot(tips, x='day', y='total_bill', kind='bar') #określa srednią i przedział ufności
```

Countpolot

```python
_ = sns.catplot(tips, x='day', kind='count') #okresla ilosc zliczen
```

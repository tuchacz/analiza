```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df_house = pd.read_csv('../../Pliki/04_seaborn/house_price_filter.csv')
df_house.head()
```

```python
df_house.info()
```

Analiza jednoczynnikowa

```python
_ = sns.catplot(df_house, x='year', kind='count')  # z jednego roku zlicza dlatego jest count
```

```python
_ = sns.catplot(df_house, x='month', kind='count')
```

```python
_ = sns.displot(df_house, x='price', bins=70)  # wyświetlenie histogramu cen, bins zgrupowanie na 70
```

```python
_ = sns.displot(df_house, x='price_per_meter', bins=70)  # cena za metr kw
```

```python
_ = sns.catplot(df_house, x='floor', kind='count')
```

```python
_ = sns.displot(df_house, x='area', bins=50)  # bins 50 ilość tych słupków
```

```python
_ = sns.catplot(df_house, x='rooms', kind='count')
```

```python
_ = sns.catplot(df_house, x='offer_type', kind='count')
```

```python
_ = sns.catplot(df_house, y='offer_type_of_building', kind='count')
```

```python
_ = sns.catplot(df_house, y='voivodeship', kind='count')
```

```python
_ = sns.catplot(df_house, y='city_name', kind='count', order=df_house.value_counts('city_name').iloc[
                                                             :20].index)  # selekcja filtrowania po pierwszych 20, sortowanie po value_counts
```

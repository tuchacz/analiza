```python
data = df_delivery
target = data.pop('DeliveredOnTime')
```

```python
data.head()
```

```python
target.head()
```

Podzia na zbir treningowy i testowy

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, random_state=5) # x cechy, y etykiety, random_state losowo wyciaga cechy i etykiety i je dzieli na testowe i treningowe
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Uczenie

Trening modelu

```python
from sklearn.linear_model import LogisticRegression

log_regression = LogisticRegression()
log_regression.fit(X_train, y_train) # trenowanie
```

```python
y_pred = log_regression.predict(X_test)
y_pred
```

Ocena - macierz pomyek

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
# 130 dostawa na czas, 5 przewidzielismy ze dotrze na czas ale nie dotarlo
_ = sns.heatmap(cm, annot=True, fmt='d')
```

Ocena - dokadno

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_test, y_pred)
acc
```

```python
# dokladnosc to 98 %
```

Ocena - peny raport

```python
from sklearn.metrics import classification_report

cr = classification_report(y_test, y_pred)
print(cr)
```
